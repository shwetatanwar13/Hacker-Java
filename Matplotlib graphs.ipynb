```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(1,10,20)
y=x**2
```

```python
%matplotlib inline
```

```python
plt.plot(x,y,'y-')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Graph")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.2,0.3,0.9,0.9])
axes2=fig.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,'r--')
axes2.plot(y,x,'y*-')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x+2,color='blue',linewidth=2)
axes.plot(x,x+4,color='yellow',lw=1,ls='--')
axes.plot(x,x+6,color='green',lw=0.5,ls='--',marker='*')
axes.plot(x,x+8,color='red',lw=1,ls=':',marker='+',markersize=7)
axes.plot(x,x+10,color='purple',lw=4,ls='--',marker='o',markersize=5,markerfacecolor="yellow")
```

```python
fig,axes = plt.subplots(figsize=(12,6))
axes.plot(x,x**5,color='orange',marker='*')
axes.set_xlim([1,2])
axes.set_ylim([1,2])
```
